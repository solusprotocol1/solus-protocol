# S4 Ledger SDK Demo

This notebook demonstrates anchoring and verifying defense logistics records using S4 Ledger SDK.

```python
from s4_sdk import S4SDK

seed = 'sEdT9vPQ4QCA4TtDSZqAGTv9ABL2uLS'
sdk = S4SDK(wallet_seed=seed, testnet=True)
record = 'NSN: 2815-01-449-5765\nNomenclature: ENGINE,DIESEL\nCondition: Serviceable\nInspection: 2026-01-20\nResult: Passed - all parameters within spec'
result = sdk.anchor_record(record_text=record, encrypt_first=True, fiat_mode=False)
print(result)
```

```python
# Verify record hash
hash_val = sdk.create_record_hash(record)
print('Record hash:', hash_val)
```
